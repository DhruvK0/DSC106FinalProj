```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
patents = 'https://www.uspto.gov/web/offices/ac/ido/oeip/taf/appl_yr.htm'

#get the raw html
r = requests.get(patents)
html = r.text

#parse the html
soup = BeautifulSoup(html, 'html.parser')

#find the table
table = soup.find('table')

#within the table, find the thead and conver it to a list
thead = table.find('thead')
thead = thead.find_all('th')
columns = [th.text for th in thead]

columns
```

```python
#within the tbody, for every tr, conver the tds to a list
tbody = table.find('tbody')
rows = tbody.find_all('tr')
data = []

for row in rows:
    tds = row.find_all('td')
    data.append([td.text for td in tds])

data
```

```python
#using the columns and data, create a dataframe
df = pd.DataFrame(data, columns=columns)
df.tail(10)
```
